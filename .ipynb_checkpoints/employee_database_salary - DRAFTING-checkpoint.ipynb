```python
# Reading csv file for now
# departments_file = pd.read_csv('/Users/diannejardinez/Desktop/sql-challenge/Data/departments.csv')
# dept_emp_file = pd.read_csv('/Users/diannejardinez/Desktop/sql-challenge/Data/dept_emp.csv')
# dept_manager_file = pd.read_csv ('/Users/diannejardinez/Desktop/sql-challenge/Data/dept_manager.csv')
# employees_file = pd.read_csv ('/Users/diannejardinez/Desktop/sql-challenge/Data/employees.csv')
salaries_file = pd.read_csv ('/Users/diannejardinez/Desktop/sql-challenge/Data/salaries.csv')
titles_file = pd.read_csv ('/Users/diannejardinez/Desktop/sql-challenge/Data/titles.csv')
```

```python
#Converting file to DataFrames 
salary_df = pd.DataFrame(salaries_file)
title_df = pd.DataFrame(titles_file)
```

```python
#Merging salary df and title df on employee number
combined_df = pd.merge(salary_df, title_df, how = 'outer', on='emp_no')

#Filtering DataFrame to show only employee number, salary, and title
salary_title_df = combined_df[['title', 'salary']]
```

```python
# Function to get the salary for each position

# Positions:
# Senior Engineer
# Assistant Engineer
# Engineer
# Manager
# Technique Leader
# Senior Staff
# Staff

def salary_dataframe(position_title):
    #Creating dataframe to get a salary from a specific position
    salary_position = salary_title_df.loc[salary_title_df['title'] == position_title]['salary']
    salary_position_df = pd.DataFrame(salary_position) 
    
    #Calculating unique counts of each salary to see the frequency of that salary
    count_salary_grp = salary_position_df.groupby('salary')
    count_salary_grp = count_salary_grp['salary'].count()
    count_salary_grp
    
    #Creating new DataFrame
    salary_range_df = pd.DataFrame(count_salary_grp)
    salary_range_df = salary_range_df.rename(columns={'salary':'frequency'})
    df = salary_range_df.reset_index()
    
    return df
```

```python
#Calculating and storing salary range for each position with salary function

salary_range_sr_eng = salary_dataframe('Senior Engineer')
#lowest salary: 40,000
#highest salary: 110,449

salary_range_ast_eng = salary_dataframe('Assistant Engineer')
#lowest salary: 40,000
#highest salary: 102,288

salary_range_eng = salary_dataframe('Engineer')
#lowest salary: 40,000
#highest salary: 110,449

salary_range_man = salary_dataframe('Manager')
#lowest salary: 40,000
#highest salary: 72,446

salary_range_tcld = salary_dataframe('Technique Leader')
#lowest salary: 40,000
#highest salary: 10,4065

salary_range_srstf = salary_dataframe('Senior Staff')
#lowest salary: 40,000
#highest salary: 129,492

salary_range_stf = salary_dataframe('Staff')
#lowest salary: 40,000
#highest salary: 129,492
```

```python
# Function to make a Histogram for each position

# Positions:
# Senior Engineer
# Assistant Engineer
# Engineer
# Manager
# Technique Leader
# Senior Staff
# Staff


def salary_histogram(data_frame, bar_color, title, savfig_path):
    #figure size
    plt.figure(figsize=(10,5))

    #creating histogram
    hist = data_frame['salary'].hist( color=bar_color, alpha=0.5, edgecolor='black' )

    #formating x and y axis and title
    plt.xlabel('Salary Range(USD)',fontsize='10')
    plt.ylabel('Frequency',fontsize='10')

    plt.title(title, fontsize='15')
    
    #saving and showing figure
    plt.savefig(savfig_path)

    plt.show()
```

```python
#Getting mean of salary through grouped position titles

group_title = salary_title_df .groupby('title')
ave_salary = group_title['salary'].mean()
ave_salary_group_title_df = pd.DataFrame(ave_salary)
#ave_salary_group_title_df

#Resetting index
ave_salary_group_title_df = ave_salary_group_title_df.reset_index()
ave_salary_group_title_df 
```

```python
#Creating a figure and axis objects with subplots
fig = plt.figure(figsize=(20,10))

x_axis = ave_salary_group_title_df ['title']
y_axis = ave_salary_group_title_df ['salary'].round()

#Setting formatting for axis labels
plt.xlabel('Position', fontsize='20')
plt.xticks(range(0,len(x_axis)),fontsize='15')

plt.yticks( fontsize='15')
plt.ylabel('Average Employee Salary (USD)', fontsize='20')

plot1 = plt.bar(x_axis, y_axis,color=['#0000FF', '#add8e6', '#FF0000', '#002366', '#228B22', '#90ee90', '#ffae42'], alpha=0.5, align='center')

#Setting Title
plt.title('Average Employee Salary per Position', fontsize='30')


plt.grid(True)


#Annotating on top of each bar
#Link to source: https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py
def autolabel(rects):
    #Attach a text label above each bar in *rects*, displaying its height
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize ='15')


autolabel(plot1)

plt.savefig(f'/Users/diannejardinez/Desktop/sql-challenge/salary_data_output/Ave Employee Salary per position.png')

plt.show()

```
